## Plot Jumlah Produk pada Masing-Masing Kategori

```python
# hitung data pada kategori
category_counts = df['category'].value_counts()

# urutkan kategori secara descending
sorted_categories = category_counts.index.tolist()

# buat plot untuk urutan yang sudah dibuat
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='category', order=sorted_categories, palette="Set2")
plt.title('Count of Products by Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Insight: diketahui bahwa banyak kategori produk pada dataset, maka perlu dilakukan penggolongan ulang.

```python

# Buat dataframe dengan nama category dan jumlah count
category_counts_df = pd.DataFrame({'Category': category_counts.index, 'Count': category_counts.values})
category_counts_df
```

Insight: Diketahui bahwa kategori perfume terdiri dari banyak variasi produk. Produk perfume juga tergolong lebih sulit di deskripsikan karena tidak ada tampilan fisik. Oleh sebab itu dengan adanya sistem rekomendasi ini, experience user peminat perfume dapat meningkat karena dapat menyesuaikan wangi yang diinginkan dengan cara masing-masing user mendeskripsikan wangi tersebut.

## Plot Jumlah produk perfume pada masing-masing brand

## Membuat Word Cloud Kolom Detail

Dengan mempertimbangkan asal dari saran yang diberikan oleh sistem, kita mempertimbangkan untuk mengolah kolom detail produk dan ingredients produk. Adapun alasan pengolahan kolom adalah sebagai berikut:
- Details: Kolom ini berisi deskripsi produk yang menggambarkan nilai yang dibawa atau kandungan utama dari produk yang ada. Dengan mengolah kolom ini maka diharapkan kita dapat sebuah acuan kata yang menggambarkan suatu produk yang diinginkan user.
- Ingredients: kolom ini berisi kandungan setiap produk yang ada di sephora, dengan mengolah data ingredients kita dapat membantu user yang concern terkait bahan-bahan yang dikandung produk.

```python
df['details'] = df['details'].astype(str)
word_cloud = WordCloud(max_words=100, background_color='white', min_font_size=10).generate(" ".join(df['details']))
img = word_cloud.to_image()

fig = plt.figure(1)
plt.imshow(img)
plt.axis('off')
plt.show()
```

## Word Cloud pada Masing-Masing Kategori

```python
# Iterate over each unique category
for category in df['category'].unique():
    # Filter DataFrame for the current category
    category_df = df[df['category'] == category]

    # Concatenate details for the current category into a single string
    category_corpus = " ".join(category_df['details'])

    # Generate word cloud for the current category
    word_cloud = WordCloud(max_words=100, background_color='white', min_font_size=10).generate(category_corpus)

    # Display the word cloud
    plt.figure(figsize=(8, 6))
    plt.imshow(word_cloud, interpolation='bilinear')
    plt.title(f'Word Cloud for Category: {category}')
    plt.axis('off')
    plt.show()
```

## Melihat word count pada kolom processed detail untuk melihat sebaran kata

Untuk meningkatkan efektivitas model kami mencoba melihat seberapa banyak kata yang sering muncul dan yang paling jarang muncul, jika hal tersebut dirasa perlu untuk dihapus, maka ditambahkan ke dalam stop word.

```python
# Menghitung kata yang paling sering muncul

# Step 1: Text Preprocessing
df_clean['preprocessing_details_category'] = df_clean['preprocessing_details_category'].apply(lambda x: str(x).split())

# Step 2: hitung kemunculan kata dalam kolom
high_all = Counter([item for sublist in df_clean['preprocessing_details_category'] for item in sublist])
high_all = pd.DataFrame(high_all.most_common(20))
high_all.columns = ['Words', 'Count']

# Step 3: tampilkan DataFrame dengan gradient background
high_all_styled = high_all.style.background_gradient(cmap='Greens')
high_all_styled
```

```python
# Menghitung kata yang paling jarang muncul

# Step 2: Counting the least common words
word_counts = Counter([item for sublist in df_clean['preprocessing_details_category'] for item in sublist])
least_common = pd.DataFrame(word_counts.most_common(), columns=['Words', 'Count'])

# Step 3: Sorting the DataFrame by Count in ascending order
least_common_sorted = least_common.sort_values(by='Count', ascending=True).head(20)

# Step 4: Displaying the DataFrame with a gradient background
least_common_styled = least_common_sorted.style.background_gradient(cmap='Greens')
least_common_styled
```
